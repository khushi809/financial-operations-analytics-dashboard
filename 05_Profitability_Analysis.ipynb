```python
retail = pd.read_csv(
    "clean_retail.csv"
)

retail.head()
```

```python
retail["Revenue"] = (
    retail["Quantity"]
    * retail["Price"]
)
```

```python
customer_profit = (
    retail.groupby("Customer ID")
    ["Revenue"]
    .sum()
    .sort_values(
        ascending=False
    )
)

customer_profit.head(10)
```

```python
plt.figure(figsize=(12,6))

customer_profit.head(10).plot(
    kind="bar"
)

plt.title(
    "Top 10 Customers by Revenue"
)

plt.ylabel("Revenue")

plt.show()
```

```python
product_profit = (
    retail.groupby("Description")
    ["Revenue"]
    .sum()
    .sort_values(
        ascending=False
    )
)

product_profit.head(10)
```

```python
plt.figure(figsize=(12,6))

product_profit.head(10).plot(
    kind="bar"
)

plt.title(
    "Top Products by Revenue"
)

plt.show()
```

```python
country_profit = (
    retail.groupby("Country")
    ["Revenue"]
    .sum()
    .sort_values(
        ascending=False
    )
)

country_profit.head(10)
```

```python
plt.figure(figsize=(12,6))

country_profit.head(10).plot(
    kind="bar"
)

plt.title(
    "Top Countries by Revenue"
)

plt.show()
```

```python
top5_share = (
    customer_profit.head(5).sum()
    /
    customer_profit.sum()
) * 100

print(
    f"Top 5 customers contribute {top5_share:.2f}% of total revenue"
)
```

```python
customer_profit.to_csv(
    "customer_profitability.csv"
)

product_profit.to_csv(
    "product_profitability.csv"
)

country_profit.to_csv(
    "country_profitability.csv"
)

print(
    "Profitability Files Saved"
)
```
